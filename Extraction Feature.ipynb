```python

tmaxs1 = np.array(max_points_s1)/fs3
tmaxs2 = np.array(max_points_s2)/fs3

def periode_P1(max_points_s1):
    if len(max_points_s1) < 2:
        return []
    
    disS1 = [max_points_s1[i + 1] - max_points_s1[i] for i in range(len(max_points_s1) - 1)]
    return disS1

def periode_P2(max_points_s2):
    if len(max_points_s2) < 2:
        return []
    
    disS2 = [max_points_s2[i + 1] - max_points_s2[i] for i in range(len(max_points_s2) - 1)]
    return disS2

min_length = min(len(max_points_s1), len(max_points_s2))
max_points_s1 = max_points_s1[:min_length]
max_points_s2 = max_points_s2[:min_length]

def Periode_P1P2(max_points_s1, max_points_s2):
    disS1S2 = [max_points_s2[i] - max_points_s1[i] for i in range(len(max_points_s1))]
    return disS1S2

def vpeak(normalize_rekon4):
    vpp=[]
    for i in range(len(normalize_rekon4)) :
        max_amplitude = np.max(normalize_rekon4[i])
        min_amplitude = np.min(abs(normalize_rekon4[i]))
    
        vpp.append( max_amplitude + min_amplitude )
        
    return vpp

vpp = vpeak(normalize_rekon2)
t2 = np.arange(len(normalize_rekon2)) /fs3

def mean(t2):
    return sum(t2)/len(t2)

def standar_dev(t2):
    rata2 = mean(t2)
    variance_sum = sum((x - rata2) ** 2 for x in t2)
    sample_variance = variance_sum / (len(t2) - 1)
    sample_standard_deviation = math.sqrt(sample_variance)
    return sample_standard_deviation

def rms(vpp):
    squared_data = np.square(vpp) 
    mean_squared = np.mean(squared_data)  
    rms = np.sqrt(mean_squared)  
    return rms

# Periode P1
p_1 = []
P1 = periode_P1(tmaxs1)
print("Periode P1 :", P1)
p_1.append(P1)
periodeS1 = np.mean(p_1)
print("Periode P1 :", periodeS1)

# Periode P2
P2 = periode_P2(tmaxs2)
print("Periode P2 :", P2)
periodeS2 = np.mean(P2)
print("Periode P2 :", periodeS2)

# Periode P1 P2
P1danP2 = Periode_P1P2(tmaxs1, tmaxs2)
formatted_values = []
for value in P1danP2:
    formatted_value = "{:.2f}".format(value)
    formatted_values.append(formatted_value)

numeric_values = [float(value) for value in formatted_values]

periodeS1S2 = np.mean(numeric_values)
print("Periode P1 dan P2:", periodeS1S2)

# mean value
rata2 = mean(vpp)
print("Rata-rata : ", rata2)

# standard deviation
standar_deviasi = standar_dev(vpp)
print("Standar deviasi :", standar_deviasi)

# Root mean square
root_mean_square = rms(vpp)
print("Root mean square :", root_mean_square)
```

```python
segmen_sinyal_list = []
plt.figure(figsize=(14, 7))

fs3 = 500

for i in range(len(tsnew1)):
    tsn1 = int(tsnew1[i] * fs3)
    if len(tsnew2) > i:
        tsn2 = int(tsnew2[i] * fs3)
        if tsn2 < len(normalize_a2rekon):
            segmen_sinyal = normalize_a2rekon[tsn1:tsn2]
            segmen_sinyal_list.append(segmen_sinyal)
                
            waktu = [t / fs3 for t in range(tsn1, tsn2)]
            plt.plot(waktu, segmen_sinyal, label=f'Segmen {i+1}')
        else:
            print(f"WARNING: tsnew2[{i}] longer than a2_rekonstruksi.")
    else:
        print("WARNING: total element in tsnew2 not enough")
    

plt.xlabel('Sampel')
plt.ylabel('Amplitudo')
plt.title('Segmen Signal between Tsnew1 and Tsnew2')
plt.legend()
plt.show()

for i, segmen_sinyal in enumerate(segmen_sinyal_list):
    print(f"Segmen {(i+1)/fs3}: {segmen_sinyal}")
    print(f"Panjang Segmen {(i+1)/fs3}: {len(segmen_sinyal)/fs3}")
    

freq_segments = []
power_spectrum_segments = []
meanfnn = []
mpf2 = []
products = []

for segmen_sinyal in segmen_sinyal_list:
    fft = np.fft.fft(segmen_sinyal)
    freq = np.fft.fftfreq(len(segmen_sinyal), 1/fs3)
    power_spectrum = np.abs(fft)

    freq_segment = freq[1:len(segmen_sinyal)//2]
    power_spectrum_segment = power_spectrum[1:len(segmen_sinyal)//2]

    freq_segments.append(freq_segment)
    power_spectrum_segments.append(power_spectrum_segment)

    total_power = max(freq_segment)
    meanfnn.append(total_power)

    product = freq_segment * power_spectrum_segment
    products.append(product)

    sum_product = np.sum(product)
    mpf = sum_product / total_power
    mpf2.append(mpf)


mean_mpf = np.mean(mpf2)
print("Rata-rata MPF:", mean_mpf)
min_length = min(len(f) for f in freq_segments)

freq_avg = np.zeros(min_length)
powerspectrumtot = np.zeros(min_length)

for freq_segment, power_spectrum_segment in zip(freq_segments, power_spectrum_segments):
    freq_avg = freq_segment[:min_length]
    powerspectrumtot += power_spectrum_segment[:min_length]

plt.plot(freq_avg, powerspectrumtot)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Total Power Spectrum')
plt.show()

total_power_combined = max(freq_avg)
product_combined = freq_avg * powerspectrumtot
sum_product_combined = np.sum(product_combined)
mpf_combined = sum_product_combined / total_power_combined

for product,mpf in zip(products,mpf2):
    sigma_xnorm = (np.abs(product - mpf))**2

meanfn=np.mean(meanfnn)
jumsum=np.sum(sigma_xnorm)
std_devs=np.sqrt(jumsum/meanfn)
print("standar deviasi:", std_devs)
```
